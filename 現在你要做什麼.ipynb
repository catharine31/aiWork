```python
import pandas as pd
df = pd.read_excel('無聊的生活.xlsx')
```

```python
df.rename(columns={'Unnamed: 0':'時間'}, inplace=True)
df
```

```python
print("輸入星期及時間（格式：星期 時間(24小時制，以整點為單位)，如：星期一 7點）")
weekAndTime = input('星期 時間：')
def getTime(weekAndTime):
    try:
        weekAndTime = weekAndTime.split(" ")
        day = weekAndTime[0]
        if day == "星期天":
            day = "星期日"
        time = weekAndTime[1].split("點")[0]
        return(day, time)
    except:
        print("請正確輸入～")
def toDo():
    time = getTime(weekAndTime)
    day = time[0]
    t = time[1]
    t = int(t) - 7
    print("你應該去", df[time[0]][t])
toDo()
```
