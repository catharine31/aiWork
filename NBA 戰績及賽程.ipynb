```python
# NBA 賽程
df_temp = pd.read_html('https://s5.sir.sportradar.com/taiwansportslottery/zht/2/season/68176/fixtures')
len(df_temp)
```

```python
df = df_temp[0]
column_name = df.loc[0, :]
df.columns = column_name
df
df.drop(0, inplace=True)
df.dropna(axis=1, how='all', inplace=True)
```

```python
df
```

```python
len(df)
```

```python
def getScore(a):
    try:
        b = a.split(':')
        c = ':'.join( [b[0], b[2]])
        return c
    except (AttributeError, IndexError):
        return "--"
```

```python
for i in range(1, len(df)):
    for j in range(2, 8):
        newScore = getScore(df.iloc[i, j])
        df.iloc[i, j] = newScore
df
```
